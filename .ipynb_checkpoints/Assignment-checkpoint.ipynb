```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"C:\Users\ADMIN\Desktop\KeSRA Data Science Training\Week 5\GIT Hub\KRA_Class_Dataset.xlsx"  # Update path accordingly

df = pd.read_excel(file_path)
```

#### Bar graph

```python
import seaborn as sns
import matplotlib.pyplot as plt

sector_tax = df.groupby('sector')['tax_paid'].sum().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=sector_tax.index, y=sector_tax.values, color='skyblue')


plt.title('Total Tax Paid per Sector')
plt.xlabel('Sector')
plt.ylabel('Total Tax Paid')


plt.xticks(rotation=45)


plt.show()
```

#### Pie chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

taxpayer_counts = df['taxpayer_type'].value_counts()


plt.figure(figsize=(6, 6))
plt.pie(taxpayer_counts, labels=taxpayer_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'grey', 'lightblue'])


plt.title('Proportion of Taxpayer Types')
plt.ylabel('')  # Remove y-axis label


plt.savefig("Pie_chart.pdf", dpi=300, bbox_inches='tight')


plt.show()
```

#### Histigram

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.histplot(df['tax_paid'], bins=20, kde=False, color='blue', edgecolor='black')

plt.title('Distribution of Tax Paid')
plt.xlabel('Tax Paid')
plt.ylabel('Frequency')


plt.show()
```
